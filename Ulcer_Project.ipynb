# Neural Network

```python

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from IPython import display
display.set_matplotlib_formats("svg")

# Load dataset
df = df_diabetic_full.copy()

lactate_columns = [col for col in df.columns if "Lactate" in col]
wound_columns = [col for col in df.columns if "wc" in col]

print(f"Lactate columns: {len(lactate_columns)}, Wound closure columns: {len(wound_columns)}")

df[lactate_columns] = df[lactate_columns].fillna(0)  # Replace NaNs in lactate data
df[wound_columns] = df[wound_columns].fillna(0)  # Replace NaNs in wound closure data


#df_filtered = df.dropna(subset=wound_columns)  # Drop rows where Wound Closure is NaN

# Extract X (Lactate) and Y (Wound Closure) for all mice
#X = df[lactate_columns].values  # Lactate as predictor

#y = df[wound_columns].values  # wound closure as targets

from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X = scaler_X.fit_transform(df[lactate_columns].values)
y = scaler_y.fit_transform(df[wound_columns].values)





# Convert to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)
print(X_tensor.shape, y_tensor.shape)

df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle dataset


# Prepare traning and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, train_size=0.8, random_state=42)

X_test = torch.nan_to_num(X_test, nan=0.0)  # Replace NaNs with zero


train_data = TensorDataset(X_train, y_train)
test_data = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_data, shuffle=True, batch_size=16)
test_loader = DataLoader(test_data, batch_size=len(test_data), shuffle=False)




 # Build a neural net model with a single hidden layer of 16 neurons
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.input = nn.Linear(in_features=10, out_features=32)  # Increased neurons
        self.bn1 = nn.BatchNorm1d(32, momentum=0.9)
        self.hidden_1 = nn.Linear(in_features=32, out_features=16)
        self.bn2 = nn.BatchNorm1d(16, momentum=0.9)
        self.hidden_2 = nn.Linear(in_features=16, out_features=16)  # Added second hidden layer
        self.dropout = nn.Dropout(p=0.3)
        self.output = nn.Linear(in_features=16, out_features=10)

    def forward(self, x):
        x = F.relu(self.bn1(self.input(x)))
        x = F.relu(self.bn2(self.hidden_1(x)))
        x = self.dropout(F.relu(self.hidden_2(x)))
        return self.output(x)

model = Net()



print(model)


# Train the neural net using the training set and then test its accuracy using the
# testing set

num_epochs = 200
train_losses, test_losses = [], []

loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=15)



for epoch in range(num_epochs):
    batch_losses = []
    for X, y in train_loader:
        preds = model(X)
        loss = loss_function(preds, y)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # Prevent unstable gradients
        optimizer.step()

        batch_losses.append(loss.item())

    train_losses.append(np.mean(batch_losses))

    # Test set
    X, y = next(iter(test_loader))
    with torch.no_grad():
        preds = model(X)
        loss = loss_function(preds, y)
        test_losses.append(loss.item())

    scheduler.step(test_losses[-1])  # Reduce learning rate dynamically




# Visualize the testing and training accuracies

fig = plt.figure(tight_layout=True)
gs = gridspec.GridSpec(nrows=2, ncols=1)

ax = fig.add_subplot(gs[0, 0])
ax.plot(train_losses)
ax.set_xlabel("Epoch")
ax.set_ylabel("Training Loss")

ax = fig.add_subplot(gs[1, 0])
ax.plot(test_losses)
ax.set_xlabel("Epoch")
ax.set_ylabel("Test Loss")

fig.align_labels()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np
import pandas as pd

# Step 1: Preprocess the data (scaling and feature engineering)
scaler = StandardScaler()

# Feature Engineering
X = regression_data[['Time', 'Lactate']].copy()
X['Time_Since_Last'] = X['Time'].diff().fillna(0)  # Time since last observation
X['Lactate_Time_Interaction'] = X['Lactate'] * X['Time']  # Interaction term

# Adding lag and rolling features
X['lactate_lag1'] = X['Lactate'].shift(1)
X['lactate_lag2'] = X['Lactate'].shift(2)
X['lactate_rolling_mean'] = X['Lactate'].rolling(window=5).mean()

# Fill NaN values appropriately
X = X.fillna(method='bfill').fillna(0)  # Backfill first, then fill remaining with 0

# Verify that X is not empty after processing
if X.empty:
    raise ValueError("Feature DataFrame X is empty after processing.")

# Align the target variable 'y' with the modified features 'X'
y = regression_data['WoundClosure'].reindex(X.index)

# Scale the features
X_scaled = scaler.fit_transform(X)

# Debugging outputs
print(f"Feature DataFrame X:\n{X.head()}")
print(f"Scaled features (first 5 rows):\n{X_scaled[:5]}")
print(f"Target variable y (first 5 rows):\n{y[:5]}")

# Step 2: Define evaluation metrics
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)  # Custom scorer for RMSE

# Step 3: Initialize the KNN model and cross-validation
knn_model = KNeighborsRegressor(metric='manhattan')

# Adjust TimeSeriesSplit for smaller datasets
tscv = TimeSeriesSplit(n_splits=min(5, len(X_scaled) // 2))  # Ensure enough samples per split

# Calculate minimum samples per fold
min_samples_per_fold = float('inf')  # Initialize with a very large value
for train_index, test_index in tscv.split(X_scaled):
    min_samples_per_fold = min(min_samples_per_fold, len(train_index))

# Debug output for minimum samples per fold
print(f"Minimum samples per fold: {min_samples_per_fold}")

# Define the parameter grid for n_neighbors
param_grid = {'n_neighbors': range(1, min(min_samples_per_fold + 1, 10))}
print(f"Parameter grid for n_neighbors: {list(param_grid['n_neighbors'])}")

# Step 4: Evaluate KNN with rolling window cross-validation
r2_scores = cross_val_score(knn_model, X_scaled, y, cv=tscv, scoring='r2')
rmse_scores = cross_val_score(knn_model, X_scaled, y, cv=tscv, scoring=rmse_scorer)

print(f"KNN Rolling Window Cross-validated R^2 scores: {r2_scores}")
print(f"Mean R^2 score: {r2_scores.mean()}")
print(f"KNN Rolling Window Cross-validated RMSE scores: {rmse_scores}")
print(f"Mean RMSE score: {abs(rmse_scores.mean())}")

# Step 5: Perform grid search for hyperparameter tuning
grid_search = GridSearchCV(KNeighborsRegressor(metric='manhattan'), param_grid, cv=tscv, scoring='r2')
grid_search.fit(X_scaled, y)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best R^2 score: {grid_search.best_score_}")

# Step 6: Train the final model with the best parameters
best_knn = KNeighborsRegressor(n_neighbors=grid_search.best_params_['n_neighbors'], metric='manhattan')
best_knn.fit(X_scaled, y)

# Final debug output
print("Final KNN model trained with best parameters.")
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer
import numpy as np

# Step 1: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Ensure X is your feature DataFrame

# Step 2: Define the custom RMSE scorer
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)

# Step 3: Initialize TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)  # Adjust n_splits as per your dataset size

# Step 4: Define Lasso model and parameter grid
lasso = Lasso(max_iter=10000, random_state=42)
param_grid = {'alpha': np.logspace(-4, 2, 50)}  # Alpha values from 0.0001 to 100

# Step 5: Perform GridSearchCV
grid_search = GridSearchCV(lasso, param_grid, cv=tscv, scoring='r2')
grid_search.fit(X, y)

# Step 6: Print results
print(f"Best alpha: {grid_search.best_params_['alpha']}")
print(f"Best R^2 score: {grid_search.best_score_}")

# Step 7: Train the final model with the best alpha
best_lasso = Lasso(alpha=grid_search.best_params_['alpha'], max_iter=10000, random_state=42)
best_lasso.fit(X, y)

# Step 8: Evaluate the final model
y_pred = best_lasso.predict(X)
final_rmse = rmse(y, y_pred)
print(f"Final RMSE: {final_rmse}")
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Create pipeline for scaling and SVR
svm_model = make_pipeline(StandardScaler(), SVR(kernel='linear'))  # Try different kernels

# Set up TimeSeriesSplit for cross-validation
tscv = TimeSeriesSplit(n_splits=5)

# Evaluate the model using cross-validation
scores = cross_val_score(svm_model, X, y, cv=tscv, scoring='neg_mean_squared_error')
r2_scores = cross_val_score(svm_model, X, y, cv=tscv, scoring='r2')

# Calculate RMSE from MSE scores
rmse_scores = np.sqrt(-scores)

# Print the evaluation metrics
print(f"Cross-validated RMSE scores: {rmse_scores}")
print(f"Mean RMSE: {rmse_scores.mean()}")
print(f"Cross-validated R^2 scores: {r2_scores}")
print(f"Mean R^2: {r2_scores.mean()}")
```
